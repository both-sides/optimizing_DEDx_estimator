```python
filename = "data/HSCPgluino_M-1800_fromAOD.root"
treename = "HSCPFullAODAnalyzer/" + dir.Get("Events").GetName()


df0 = rt.RDataFrame(treename, filename)
```

```python
df0.Display().Print()
```

```python
df = df0.Filter(
    "HLT_Mu50 && ROOT::VecOps::Sum(IsoTrack_pt > 55) > 0",
    "HLT_Mu50_and_anyIsoTrack55"
)
```

```python
df.Display().Print()
```

```python
hmncSbr = ["DeDx_IhStrip1","DeDx_IhStrip", "DeDx_IhStrip3", "DeDx_IhStrip4"]
hists1 = {}
truncSbr = ["DeDx_ItStrip0","DeDx_ItStrip5", "DeDx_ItStrip10", "DeDx_ItStrip15", "DeDx_ItStrip20", "DeDx_ItStrip25", "DeDx_ItStrip30", "DeDx_ItStrip35", "DeDx_ItStrip"]
hists2 = {}
```

```python

color_map = {
    1: rt.kViolet,
    2: rt.kPink,
    3: rt.kGray,
    4: rt.kGreen,
    5: rt.kBlue,
    6: rt.kYellow,
    7: rt.kMagenta,
    8: rt.kCyan,
    9: rt.kOrange,
    10: rt.kBlack
}
```

```python
%jsroot
c = rt.TCanvas("c","Log-scale overlay",800,600)
c.SetLogy()

leg = rt.TLegend(0.65,0.52,0.88,0.7)

for i, branch in enumerate(hmncSbr):
    if i == 1:
        hists1[f"h{i-1}"].SetTitle("Generalized harmonic mean")
        hists1[f"h{i-1}"].GetXaxis().SetTitle("DEDx_lhStrip(GEV)")
    hists1[f"h{i}"] = df0.Histo1D((f"h{i}","",100, 0, 40),branch)
    hists1[f"h{i}"].SetLineColor(color_map[i+1])
    hists1[f"h{i}"].Draw("same")
    leg.AddEntry( hists1[f"h{i}"].GetPtr(), f"h{i}", "l" )
    
max_vals = [h.GetMaximum() for key,h in hists1.items()]
overall_max = max(max_vals)
hists1["h0"].SetMaximum(overall_max * 1.1)


leg.Draw()

c.Draw()
```

```python
%jsroot
c = rt.TCanvas("c","Log-scale overlay",800,600)
c.SetLogy()

leg = rt.TLegend(0.65,0.52,0.88,0.7)

for i, branch in enumerate(hmncSbr):
    if i == 1:
        hists1[f"h{i-1}"].SetTitle("Generalized harmonic mean (HLT_Mu50 with Isotrack_pt>55)")
        hists1[f"h{i-1}"].GetXaxis().SetTitle("DEDx_lhStrip(GEV)")
    hists1[f"h{i}"] = df.Histo1D((f"h{i}","",100, 0, 40),branch)
    hists1[f"h{i}"].SetLineColor(color_map[i+1])
    hists1[f"h{i}"].Draw("same")
    leg.AddEntry( hists1[f"h{i}"].GetPtr(), f"h{i}", "l" )
    
max_vals = [h.GetMaximum() for key,h in hists1.items()]
overall_max = max(max_vals)
hists1["h0"].SetMaximum(overall_max * 1.1)


leg.Draw()

c.Draw()
```

```python
%jsroot
d = rt.TCanvas("d","Log-scale overlay",800,600)
d.SetLogy()

leg = rt.TLegend(0.65,0.52,0.88,0.7)

for i, branch in enumerate(truncSbr):
    if i == 1:
        hists2[f"h{i-1}"].SetTitle("Truncated mean")
        hists2[f"h{i-1}"].GetXaxis().SetTitle("DEDx_ItStrip(GEV)")
    hists2[f"h{i}"] = df0.Histo1D((f"h{i}", "", 100, 0, 40), branch)
    hists2[f"h{i}"].SetLineColor(color_map[i+1])
    hists2[f"h{i}"].Draw("same")
    leg.AddEntry( hists2[f"h{i}"].GetPtr(), f"h{i}", "l" )

max_vals = [h.GetMaximum() for key,h in hists2.items()]
overall_max = max(max_vals)
hists2["h1"].SetMaximum(overall_max * 1.1)

leg.Draw()
d.Draw()
```
